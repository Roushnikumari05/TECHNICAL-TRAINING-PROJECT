```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Create Synthetic Data
data = {
    'Faculty_Name': ['Dr. Smith', 'Prof. Jones', 'Dr. Taylor', 'Dr. Brown', 'Prof. White', 'Dr. Green'],
    'Department': ['CS', 'Math', 'CS', 'Physics', 'Math', 'Physics'],
    'Lecture_Hours': [12, 15, 8, 10, 18, 9],
    'Lab_Hours': [4, 2, 6, 2, 0, 4],
    'Admin_Hours': [2, 5, 2, 4, 1, 8]
}

df = pd.DataFrame(data)

# 2. Analyze Teaching Hours & Workload Balance
# Calculate Total Hours
df['Total_Hours'] = df['Lecture_Hours'] + df['Lab_Hours'] + df['Admin_Hours']

# Define Workload Balance (Threshold: > 18 hours is 'Overloaded', else 'Balanced')
df['Status'] = df['Total_Hours'].apply(lambda x: 'Overloaded' if x > 18 else 'Balanced')

# 3. Summary Statistics
print("--- Faculty Workload Summary Statistics ---")
summary = df.describe()
print(summary)
print("\n--- Workload Status per Faculty ---")
print(df[['Faculty_Name', 'Total_Hours', 'Status']])

# 4. Visualization: Bar Plot
plt.figure(figsize=(10, 6))
colors = ['red' if x > 18 else 'skyblue' for x in df['Total_Hours']]

plt.bar(df['Faculty_Name'], df['Total_Hours'], color=colors)
plt.axhline(y=18, color='gray', linestyle='--', label='Overload Threshold (18 hrs)')

plt.title('Total Weekly Workload per Faculty Member')
plt.xlabel('Faculty Name')
plt.ylabel('Total Hours')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
